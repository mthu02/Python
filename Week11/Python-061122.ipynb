```python
import glob 
import pandas as pd
import matplotlib.pyplot as plt
path = 'C:/Users/ADMIN/Desktop/Python/data'
files = glob.glob(path+"/*.txt")
df_ = (pd.read_csv(file) for file in files)
df = pd.concat(df_)
df
```

```python
def plot(df,attr):
    fig = plt.figure(figsize=(10,4))

    ax1 = fig.add_subplot(121)
    ax1.hist(df)
    ax1.set_title('Histogram about ({})'.format(attr))

    dfxx = df.describe().to_frame().round(2)

    ax2 = fig.add_subplot(122)
    font_size=12
    bbox=[0, 0, 1, 1]
    ax2.axis('off')
    mpl_table = ax2.table(cellText = dfxx.values, rowLabels = dfxx.index, bbox=bbox, colLabels=dfxx.columns)
    mpl_table.auto_set_font_size(False)
    mpl_table.set_fontsize(font_size)
```

```python
dfx=[]
for c in columns:
    d = df[df["Parameter"]==c]["Value"]
    dfx.append(d)
```

```python
idx = 0
for dfx_ in dfx:
    plot(dfx_, columns[idx])
    idx+=1
```
